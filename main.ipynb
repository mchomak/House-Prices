### Самые важные параметры влияющие на Итоговую цену дома:

- OverallQual: Общее качество материала и отделки
- YearBuilt: Первоначальная дата постройки
- YearRemodAdd: Дата реконструкции
- TotalBsmtSF: Общая площадь подвала в квадратных футах
- 1stFlrSF: Площадь первого этажа в квадратных футах
- GrLivArea: Жилая площадь выше уровня земли, квадратные футы
- TotRmsAbvGrd: Общее количество номеров выше класса (не включая ванные комнаты)
- GarageYrBlt: Год постройки гаража
- GarageArea: Размер гаража в квадратных футах
- PoolQC: Качество бассейна


```python
def preprocessing(df):
    df['Alley'].fillna('Missing', inplace=True)
    df['MasVnrType'].fillna('Missing', inplace=True)
    df['BsmtQual'].fillna('Missing', inplace=True)
    df['BsmtCond'].fillna('Missing', inplace=True)
    df['BsmtExposure'].fillna('Missing', inplace=True)
    df['BsmtFinType1'].fillna('Missing', inplace=True)
    df['BsmtFinType2'].fillna('Missing', inplace=True)
    df['Electrical'].fillna('Missing', inplace=True)
    df['FireplaceQu'].fillna('Missing', inplace=True)
    df['GarageType'].fillna('Missing', inplace=True)
    df['GarageFinish'].fillna('Missing', inplace=True)
    df['GarageQual'].fillna('Missing', inplace=True)
    df['GarageCond'].fillna('Missing', inplace=True)
    df['PoolQC'].fillna('Missing', inplace=True)
    df['Fence'].fillna('Missing', inplace=True)
    df['MiscFeature'].fillna('Missing', inplace=True)
    df['LotFrontage'].fillna('Missing', inplace=True)    

    # если не указан тип шпона, то ставим 0, а если указан, то ставим среднее значение
    # df['MasVnrArea'].fillna(lambda x: 0 if x["MasVnrType"]=="Missing" else df['MasVnrArea'].mean(), inplace=True)
    # df['GarageYrBlt'].fillna(lambda x: 0 if x["GarageType"]=="Missing" else df['GarageYrBlt'].mean(), inplace=True)
    # df['GarageCars'].fillna(lambda x: 0 if x["GarageType"]=="Missing" else df['GarageCars'].mean(), inplace=True)
    # df['GarageArea'].fillna(lambda x: 0 if x["GarageType"]=="Missing" else df['GarageArea'].mean(), inplace=True)

    df['MasVnrArea'].fillna(df['MasVnrArea'].mean(), inplace=True)
    df['GarageYrBlt'].fillna(df['GarageYrBlt'].mean(), inplace=True)
    df['GarageCars'].fillna(df['GarageCars'].mean(), inplace=True)
    df['GarageArea'].fillna(df['GarageArea'].mean(), inplace=True)

    zeros=[
        "MiscVal",
        "PoolArea",
        "ScreenPorch",
        "3SsnPorch",
        "EnclosedPorch",
        "OpenPorchSF",
        "WoodDeckSF",
        "TotalBsmtSF",
        "LowQualFinSF",
        "BsmtUnfSF",
        "BsmtFinSF2",
        "BsmtFinSF1",
        "MasVnrArea",
    ]
    for name in zeros:
        df.drop(name, axis=1, inplace=True)
    
    columns=[]
    for col in df:
        if df[col].dtypes=="object":
            columns.append(col)

    df=pd.get_dummies(df, columns=columns, drop_first= True)
    return df
    
```

```python
abt = df.copy()
abt=preprocessing(abt) 
abt.head()
```

### Setup ML Pipelines

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

```python
pipelines = {
    'rf': make_pipeline(StandardScaler(), RandomForestClassifier(random_state=123)),
    # 'gb': make_pipeline(StandardScaler(), GradientBoostingClassifier(random_state=123))
}
```

```python
GradientBoostingClassifier().get_params()
```

```python
grid = {
    'rf': {
        'randomforestclassifier__n_estimators':[100,200,300]
    }
}
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score
import pickle
import numpy as np
```

```python
def train_model(pipelines):
  fit_models = {}
  for algo, pipeline in pipelines.items():
    print(f'Training the {algo} model.')
    model = GridSearchCV(pipeline, grid[algo], n_jobs=-1, cv=10)
    model.fit(X_train, y_train)
    fit_models[algo] = model 
    with open(f'{algo}.pkl', 'wb') as f: 
      pickle.dump(fit_models[algo], f)

  return fit_models
```
